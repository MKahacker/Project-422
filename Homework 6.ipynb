```python
import numpy as np
import pandas as pd

from sklearn import metrics
```

```python
users = pd.read_csv('ml-100k/u.user', sep='|', names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])
```

```python
users.head()
```

```python
ratings = pd.read_csv('ml-100k/u.data',
                      sep='\t',
                      names=['user_id',
               'movie id',
               'rating',
               'timestamp'])
```

```python
ratings[ratings['user_id'] == 1].head()
```

```python
items = pd.read_csv('ml-100k/u.item',
                    sep='|',
                    names=['movie id',
             'movie title',
             'release date',
             'video release date',
             'IMDb URL',
             'Unknown',
             'Action',
             'Adventure',
             'Animation',
             'Childrens',
             'Comedy',
             'Crime',
             'Documentary',
             'Drama',
             'Fantasy',
             'FilmNoir',
             'Horror',
             'Musical',
             'Mystery',
             'Romance',
             'SciFi',
             'Thriller',
             'War',
             'Western'],
                    encoding='latin-1')
```

```python
items[items["movie id"] == 20]
```

```python
utility = ratings.pivot(index="user_id",
                       columns="movie id",
                       values="rating")
utility.head()
```

```python
mean_ratings = utility.mean(axis=1)
mean_ratings
```

```python
centered_u = utility - mean_ratings
centered_u = centered_u.where((pd.notnull(centered_u)),0)
centered_u.head()
```

## Problem 1
The recommender system would recommend the movie to user 50. This is because the similarity between the item and user profiles are relatively high compared to user 200. In addition item 95 is closer in distance to user 50. 

```python
u200 = centered_u.iloc[199:200,]
u200.head()
```

```python
u50 = centered_u.iloc[49:50,]
u50.head()
```

```python
users_ratings = ratings[(ratings["user_id"] == 200) | (ratings["user_id"] == 50)]
users_ratings.head()
```

```python
users_rating_feature = pd.merge(users_ratings.drop(columns=["timestamp"]), items.drop(columns=["release date", "video release date", "IMDb URL"]), on="movie id", how="inner")
users_rating_feature.head()
```

```python
u200 = users_rating_feature[users_rating_feature["user_id"] == 200]
u200.head()
```

```python
u50 = users_rating_feature[users_rating_feature["user_id"] == 50]
u50.head()
```

```python
u200["rating"].mean() 
```

```python
u50["rating"].mean()
```

```python
mean_ratings.iloc[49:50].values
```

```python
u200["rating"] = u200["rating"] - u200["rating"].mean()
```

```python
center_u50 = pd.DataFrame(u50.values, columns=u50.columns)
center_u50["rating"] = center_u50["rating"] - center_u50["rating"].mean()
center_u50
```

```python
t = []
for tcol in items.drop(columns=["movie id","movie title","release date", "video release date", "IMDb URL"]).columns:
    avg_rating = u200[u200[tcol] == 1]["rating"].mean()
    t.append([200, tcol,avg_rating])
tmp1 = pd.DataFrame(data=np.array(t), columns=["user_id","feature","rating"])
tmpu200 = tmp1.pivot(index="user_id",
                       columns="feature",
                       values="rating")
tmpu200
```

```python
t = []
for tcol in items.drop(columns=["movie id","movie title","release date", "video release date", "IMDb URL"]).columns:
    avg_rating = center_u50[center_u50[tcol] == 1]["rating"].mean()
    t.append([50, tcol,avg_rating])
tmp1 = pd.DataFrame(data=np.array(t), columns=["user_id","feature","rating"])
tmpu50 = tmp1.pivot(index="user_id",
                       columns="feature",
                       values="rating")
tmpu50
```

```python
movie95 = items.iloc[94:95]
```

```python
movie95
```

```python
print("User 50")
metrics.pairwise.cosine_similarity(movie95.drop(columns=["movie title", "release date", "IMDb URL","video release date", "movie id" ]).values, tmpu50.replace("nan", 0).values)
```

```python
print("User 200")
metrics.pairwise.cosine_similarity(movie95.drop(columns=["movie title", "release date", "IMDb URL","video release date", "movie id" ]).values, tmpu200.replace("nan", 0).values)
```

```python
print("User 50")
metrics.pairwise.cosine_distances(movie95.drop(columns=["movie title", "release date", "IMDb URL","video release date", "movie id" ]).values, tmpu50.replace("nan", 0).values)
```

```python
print("User 200")
metrics.pairwise.cosine_distances(movie95.drop(columns=["movie title", "release date", "IMDb URL","video release date", "movie id" ]).values, tmpu200.replace("nan", 0).values)
```

Movie 95 will be recommended to user 50. 

## Problem 2
The expected rating for item 508 based on the top 10 most similar users to user 1 is 3.87. 

```python
user1 = centered_u.iloc[0:1]
user1
```

```python
counter = []
for i in range(2, centered_u.shape[0]):
    u = centered_u.iloc[i-1:i]
    if metrics.pairwise.cosine_similarity(user1.values, u.values) > 0.252:
        counter.append(u)
```

```python
ph = pd.DataFrame(columns=counter[0].columns)
for i in range(len(counter)):
    ph = pd.concat([ph,counter[i]])
    
ph.describe()
```

```python
center_avg = ph[508].mean()
```

```python
the_mean = ratings[ratings["user_id"] == 1]["rating"].mean()
```

```python
expected_rating = center_avg + the_mean
expected_rating
```
