```python
rating_cols = ['user_id',
               'movie_id',
               'rating',
               'timestamp']

ratings = pd.read_csv('ml-100k/u.data',
                      sep='\t',
                      names=rating_cols)
```

```python
ratings.head()
```

```python
item_cols = ['movie id',
             'movie title',
             'release date',
             'video release date',
             'IMDb URL',
             'Unknown',
             'Action',
             'Adventure',
             'Animation',
             'Childrens',
             'Comedy',
             'Crime',
             'Documentary',
             'Drama',
             'Fantasy',
             'FilmNoir',
             'Horror',
             'Musical',
             'Mystery',
             'Romance',
             'SciFi',
             'Thriller',
             'War',
             'Western']

items = pd.read_csv('ml-100k/u.item',
                    sep='|',
                    names=item_cols,
                    encoding='latin-1')
```

```python
items.head()
```

```python
item1 = items[0:1]

item1.head()
```

```python
item2 = items[94:95]

item2.head()
```

```python
feat1 = item1.iloc[:,5:24]

feat1.head()
```

```python
feat2 = item2.iloc[:,5:24]

feat2.head()
```

```python
print(metrics.jaccard_similarity_score(feat1,feat2))
print(1 - metrics.jaccard_similarity_score(feat1,feat2))

print(metrics.pairwise.cosine_similarity(feat1,feat2))
print(metrics.pairwise.cosine_distances(feat1,feat2))
```

```python
utility = ratings.pivot(index='user_id',
                        columns='movie_id',
                        values='rating')

utility.head()
```

```python
utility.iloc[194:197,240:243]
```

```python
user_means = utility.mean(axis=1)

user_means.head()
```

```python
utility_centered = utility - user_means
utility_centered = utility_centered.where((pd.notnull(utility_centered)),0)

utility_centered.head()
```

```python
user = utility_centered.iloc[:1,]

user.head()
```
